```python
# EDA
# data cleaning
import pandas as pd
```

```python
df = pd.read_excel('/content/Customer Call List.xlsx')
df
```

```python
df=df.drop_duplicates()
df
```

```python
df['Last_Name']=df['Last_Name'].str.strip("._/")
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]',"",regex=True)
```

```python
#Standard ----> 123-545-5421
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]',"",regex=True)
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str)
df['Phone_Number'] = df['Phone_Number'].str.replace(r'\D', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)
df
```

```python
df.drop(columns='Not_Useful_Column',inplace=True)
df
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('N/a','')
df
#
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('NaN','')
df.fillna("")
```
